```python
from fst.io import read
import matplotlib.pyplot as plt
import numpy as np
import dask.array as da
%matplotlib inline
```

```python
sample = da.from_array(read('/nrs/cosem/davis/s3_testing/macrophage.n5/fibsem/s3/')).compute()
```

```python
sample.shape
```

```python
percs = np.percentile(sample, (2,50,98), axis=(1,2))
```

```python
sample.min()
```

```python
fig, axs = plt.subplots(nrows=2, squeeze=False, figsize=(12,12), dpi=200)

test_im = sample[:,:, sample.shape[2]//2].T

axs[0,0].imshow(test_im, clim=np.percentile(test_im, (2,98)), cmap='gray')
[axs[1,0].plot(p) for p in percs]
```
